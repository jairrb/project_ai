```python
df = pd.DataFrame(documents)
df.columns = ['Titulo','Conteudo']
df
```

```python
model = 'models/embedding-001'
```

```python
def embed_fn(title,text):
  return genai.embed_content(model=model,
                                 content=text,
                                 title=title,
                                 task_type='RETRIEVAL_DOCUMENT'
                                 )['embedding']
```

```python
df['Embeddings'] = df.apply(lambda row: embed_fn(row['Titulo'],row['Conteudo']), axis=1)
df
```

```python
def generate_and_query(query, base, model):
  embed_query = genai.embed_content(model=model,
                                    content=query,
                                    task_type='RETRIEVAL_QUERY')['embedding']
  product = np.dot(np.stack(df['Embeddings'] ),embed_query)

  index = np.argmax(product)
  return df.iloc[index]['Conteudo']
```

```python
consult = 'Como faço para trocar marchas em um carro do Google?'
trecho = generate_and_query(consult, df, model)
print(trecho)
```

```python
generation_config = {
    'temperature':0.5,
    'candidate_count':1,
}
```

```python
prompt = f'Reescreva esse texto de uma forma mais descontraída, sem adicionar informações que não façam parte do texto: {trecho}'

model_2 = genai.GenerativeModel('gemini-1.0-pro',
                                generation_config=generation_config)
response = model_2.generate_content(prompt)
print(response.text)
```
